```python
image_directory = 'C:/SachinP/Dataset/RFMiD/Evaluation_Set/Validation/'
image_directory_r='C:/SachinP/Dataset/RFMiD/Evaluation_Set/Validation_RGB/Val_r/'
image_directory_g='C:/SachinP/Dataset/RFMiD/Evaluation_Set/Validation_RGB/Val_g/'
image_directory_b='C:/SachinP/Dataset/RFMiD/Evaluation_Set/Validation_RGB/Val_b/'
SIZE = 299

image_dataset = []   
images = os.listdir(image_directory)
print(len(images))
for i, image_name in enumerate(images):    #Remember enumerate method adds a counter and returns the enumerate object
    if (image_name.split('.')[1] == 'png'):
        print(i)
        #print(image_directory+image_name)
        image = cv2.imread(image_directory+image_name,1)
        #image = Image.fromarray(image)
        image = cv2.resize(image,(SIZE, SIZE))
        hist,bins = np.histogram(image.flatten(),256,[0,256])
        cdf = hist.cumsum()
        cdf_normalized = cdf * float(hist.max()) / cdf.max() 
        cdf_m = np.ma.masked_less(cdf,0)
        cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
        cdf = np.ma.filled(cdf_m,0).astype('uint8')       
        image_equalize= cdf[image]
        print(image_equalize.shape)
        #image_equalize=cv2.cvtColor(image_equalize, cv2.COLOR_BGR2GRAY)
        b, g, r = cv2.split(image_equalize)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(32,32))
        image_clahe_r=clahe.apply(r)
        image_clahe_g=clahe.apply(g)
        image_clahe_b=clahe.apply(b)
        img_r=image_clahe_r
        img_g=image_clahe_g
        img_b=image_clahe_b
        filename_r = image_directory_r+str(i+8000+1)+'_r.png'
        filename_g = image_directory_g+str(i+8000+1)+'_g.png'
        filename_b = image_directory_b+str(i+8000+1)+'_b.png'
        
        # Using cv2.imwrite() method
        # Saving the image
        cv2.imwrite(filename_r, img_r)
        cv2.imwrite(filename_g, img_g)
        cv2.imwrite(filename_b, img_b)
        image_clahe_r=np.expand_dims((np.array(image_clahe_r)),2)
        image_clahe_g=np.expand_dims((np.array(image_clahe_g)),2)
        image_clahe_b=np.expand_dims((np.array(image_clahe_b)),2) 
        image_clahe= Concatenate(axis=-1)([image_clahe_r,image_clahe_g,image_clahe_b])
        
        lab = cv2.merge([image_clahe_r,image_clahe_g,image_clahe_b])
        bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
        #filename_b = image_directory_d+str(i+8000+1)+'_t.png'
        #cv2.imwrite(filename_b, bgr)
        
        image_dataset.append(np.array(image_clahe))
        
```

```python
import cv2
import numpy as np

# load image
img = cv2.imread('C:/SachinP/Dataset/RFMiD/Training_Set/Training_Set/Training_set_FD_S_SMOTE_Resize/ARMD_Rename/7000_b.png')

# convert to graky
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# threshold input image as mask
mask = cv2.threshold(gray, 0, 20, cv2.THRESH_BINARY)[1]

# negate mask
mask = 255 - mask

# apply morphology to remove isolated extraneous noise
# use borderconstant of black since foreground touches the edges
kernel = np.ones((3,3), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# anti-alias the mask -- blur then stretch
# blur alpha channel
mask = cv2.GaussianBlur(mask, (0,0), sigmaX=2, sigmaY=2, borderType = cv2.BORDER_DEFAULT)

# linear stretch so that 127.5 goes to 0, but 255 stays 255
mask = (2*(mask.astype(np.float32))-255.0).clip(0,255).astype(np.uint8)

# put mask into alpha channel
result = img.copy()
result = cv2.cvtColor(result, cv2.COLOR_BGR2BGRA)
result[:, :, 3] = mask

# save resulting masked image
cv2.imwrite('C:/SachinP/Dataset/RFMiD/Training_Set/Training_Set/Training_set_FD_S_SMOTE_Resize/DN_Rename/7000_bb.png', result)

# display result, though it won't show transparency
cv2.imshow("INPUT", img)
cv2.imshow("GRAY", gray)
cv2.imshow("MASK", mask)
cv2.imshow("RESULT", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
